```python
string_columns =[]
for col in df_new.select_dtypes(include=['object']).columns:
    string_columns.append(col)

string_columns          
```

```python
numeric_columns = []
for col in df_new.select_dtypes(include=['float']).columns:
    numeric_columns.append(col)

numeric_columns 
```

```python

#for col in numeric_columns:
    #df_new[col].fillna(df_new[col].mean, inplace=True)

for col in string_columns:
    df_new[col].fillna(df_new[col].mode()[0], inplace=True)   

#df_new.isnull().sum(axis = 0)
```

```python
for col in string_columns:
    print(col,df_new[col].isnull().sum())
```

```python
for col in numeric_columns:
    df_new[col].fillna(df_new[col].median(), inplace=True)
```
