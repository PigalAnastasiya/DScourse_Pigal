### Попробуйте различные расписания скорости обучения (например, экспоненциальное уменьшение) и сравните результаты.

Использование экспоненциального распада для снижения скорости обучения в процессе обучения модели. Точность уменьшилась на 0.004

```python
# Импортируем необходимые библиотеки
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from keras.callbacks import LearningRateScheduler

# Функция затухания шага
def scheduler(epoch, lr):
    if epoch < 10:
        return lr
    else:
        return lr * tf.math.exp(-0.1)

# Загрузим данные
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Нормализуем данные
train_images = train_images / 255.0
test_images = test_images / 255.0

# Создаем модель
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10)
])

# Компилируем модель
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
callback = LearningRateScheduler(scheduler)
# Обучаем модель
model.fit(train_images, train_labels, epochs=10,callbacks=[callback])

# Оценим точность модели
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nТестовая точность:', test_acc)
```

Использование пошагового распада(Step Decay) для снижения скорости обучения и увеличение эпох до 30. Точность увеличилась на 0.0011

```python
# Импортируем необходимые библиотеки
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from keras.callbacks import LearningRateScheduler

# Функция затухания шага(снижение скорости обучения наполовину каждые 10 эпох )
def step_decay(epoch):
    initial_lrate = 0.001
    drop = 0.1
    epochs_drop = 4.0
    lrate = initial_lrate * np.power(drop, np.floor((epoch) / epochs_drop))
    return lrate

# Загрузим данные
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Нормализуем данные
train_images = train_images / 255.0
test_images = test_images / 255.0

# Создаем модель
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10)
])

#Создаем пользовательский обратный вызов
class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
       self.losses = []
       self.lr = []
 
    def on_epoch_end(self, batch, logs={}):
       self.losses.append(logs.get('loss'))
       self.lr.append(step_decay(len(self.losses)))


loss_history = LossHistory()

lrate = LearningRateScheduler(step_decay)
callbacks_list = [loss_history, lrate]

# Компилируем модель
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
# Обучаем модель
model.fit(train_images, train_labels, epochs=30,callbacks=[callbacks_list])

# Оценим точность модели
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nТестовая точность:', test_acc)
```
